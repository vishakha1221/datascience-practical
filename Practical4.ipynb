```python
#1.	Normal distribution

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, poisson, expon, skew, kurtosis

# Generate 1000 random numbers with mean=0, std=1
normal_dist = np.random.normal(0, 1, 1000)

# Calculate mean and standard deviation
mean = np.mean(normal_dist)
std_dev = np.std(normal_dist)

# Plot PDF and CDF
x = np.linspace(min(normal_dist), max(normal_dist), 100)
plt.figure(figsize=(12,5))

# Histogram (PDF approximation)
plt.subplot(1,2,1)
plt.hist(normal_dist, bins=30, density=True, alpha=0.5, label="Data histogram")
plt.plot(x, norm.pdf(x, mean, std_dev), 'r-', lw=2, label="PDF")
plt.title("Normal Distribution - PDF")
plt.legend()

# CDF
plt.subplot(1,2,2)
plt.plot(x, norm.cdf(x, mean, std_dev), 'g-', lw=2, label="CDF")

plt.title("Normal Distribution - CDF")

plt.legend()
plt.show()
```

```python
#2.	Poisson Distribution

# Generate 1000 random numbers with =5
poisson_dist = np.random.poisson(5, 1000)
# Calculate mean and variance
poisson_mean = np.mean(poisson_dist)
poisson_var = np.var(poisson_dist)

# Plot PDF and CDF
x = np.arange(0, 15)
plt.figure(figsize=(12,5))

# Histogram (PDF approximation)
plt.subplot(1,2,1)
plt.hist(poisson_dist, bins=15, density=True, alpha=0.5, label="Data histogram")
plt.plot(x, poisson.pmf(x, 5), 'r-', lw=2, label="PMF")
plt.title("Poisson Distribution - PMF")
plt.legend()
# CDF
plt.subplot(1,2,2)
plt.plot(x, poisson.cdf(x, 5), 'g-', lw=2, label="CDF")
plt.title("Poisson Distribution - CDF")
plt.legend()
plt.show()
```

```python
#Statistics

print("Normal Distribution -> Mean:", mean, "Std Dev:", std_dev)
print("Poisson Distribution -> Mean:", poisson_mean, "Variance:", poisson_var)
```

```python
#3.	Exponential Distribution

#  = 1.0 (scale = 1/ = 1.0), size = 1000
exp_dist = np.random.exponential(scale=1.0, size=1000)

# Descriptive statistics
mean = np.mean(exp_dist)
variance = np.var(exp_dist)
skewness = skew(exp_dist)
kurt = kurtosis(exp_dist)
print("Exponential Distribution (=1)")
print("Mean:", mean)
print("Variance:", variance)
print("Skewness:", skewness)
print("Kurtosis:", kurt)

# Plot PDF and CDF
x = np.linspace(0, np.max(exp_dist), 100)
plt.figure(figsize=(12,5))

# Histogram + PDF
plt.subplot(1,2,1)
plt.hist(exp_dist, bins=30, density=True, alpha=0.5, label="Data histogram")
plt.plot(x, expon.pdf(x, scale=1.0), 'r-', lw=2, label="PDF")
plt.title("Exponential Distribution - PDF")
plt.legend()

# CDF
plt.subplot(1,2,2)
plt.plot(x, expon.cdf(x, scale=1.0), 'g-', lw=2, label="CDF")
plt.title("Exponential Distribution - CDF")
plt.legend()
plt.show()
```
